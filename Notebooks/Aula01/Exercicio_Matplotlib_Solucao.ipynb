```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../../Datasets/german_credit_data.csv')
data
```

```python
# Gráfico de Linha: Variação de idades ao longo dos registros
plt.figure(figsize=(10, 6))
plt.plot(data['Age'])
plt.title('Variação de Idades')
plt.xlabel('Registro')
plt.ylabel('Idade')
plt.show()
```

```python
data['Sex'].value_counts()
```

```python
# Gráfico de Barras: Contagem de Pessoas por Sexo

sex_counts = data['Sex'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(sex_counts.index, sex_counts.values, color='skyblue')
plt.title('Contagem de Pessoas por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.show()
```

```python
# Gráfico de Pizza: Distribuição dos propósitos
plt.figure(figsize=(8, 8))
plt.pie(data['Purpose'].value_counts(), labels=data['Purpose'].unique(), autopct='%1.1f%%', startangle=140, textprops={'color':"white"})
plt.title('Distribuição dos propósitos', color='white')
plt.show()
```

```python
# Histograma: Distribuição de Idades
plt.figure(figsize=(10, 6))
plt.hist(data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição de Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de Dispersão: Idade vs. Valor do Crédito
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Credit amount'], alpha=0.5)
plt.title('Gráfico de Dispersão: Idade vs. Valor do Crédito')
plt.xlabel('Idade')
plt.ylabel('Valor do Crédito')
plt.show()
```

```python
# Gráfico de Boxplot: Distribuição de Idade por Sexo
plt.figure(figsize=(8, 6))
plt.boxplot([data[data['Sex'] == 'male']['Age'], data[data['Sex'] == 'female']['Age']], labels=['Masculino', 'Feminino'])
plt.title('Comparação de Idade por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Idade')
plt.show()
```
